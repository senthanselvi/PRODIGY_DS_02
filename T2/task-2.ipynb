```python
import pandas as pd

# Load the Titanic dataset
titanic_df = pd.read_csv('/kaggle/input/titanic/train.csv')  # Replace with the actual file path if different
titanic_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.DataFrame(titanic_df)
```

```python
print("First few rows of the dataset:")
print(titanic_df.head())
```

```python
# Check for missing values in the dataset
print(titanic_df.isnull().sum())

# Fill missing values in 'Age' with the median age
#titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

titanic_df.fillna({'Age' : titanic_df['Age'].median()})

# Fill missing values in 'Embarked' with the most frequent value
#titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

titanic_df.fillna({'Embarked' : titanic_df['Embarked'].mode()[0]})

# Drop the 'Cabin' column due to a large number of missing values
titanic_df.drop(columns=['Cabin'], inplace=True)

# Verify that there are no more missing values
print(titanic_df.isnull().sum())
```

```python
print("\nSummary statistics:")
print(titanic_df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survival Count by Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'], kde=True, bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of 'Fare'
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(titanic_df[['Age', 'Fare', 'Survived']], hue='Survived', diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Embarked', data=titanic_df)
plt.title('Survival Count by Embarked Location')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Convert 'Sex' into numerical values (0 for female, 1 for male)
titanic_df['Sex'] = titanic_df['Sex'].map({'female': 0, 'male': 1})

# Convert 'Embarked' into numerical values using one-hot encoding
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'], drop_first=True)

# Display the first few rows to verify the changes
print(titanic_df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation heatmap
numeric_cols = titanic_df.select_dtypes(include=['float64', 'int64']).columns

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
 # Distribution of survival by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival Distribution by Gender')
plt.show()
```

```python
# Distribution of survival by passenger class
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survival Distribution by Passenger Class')
plt.show()
```

```python
# Distribution of age by survival status
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df[titanic_df['Survived'] == 1]['Age'], bins=30, kde=False, color='green', label='Survived')
sns.histplot(titanic_df[titanic_df['Survived'] == 0]['Age'], bins=30, kde=False, color='red', label='Not Survived')
plt.title('Age Distribution by Survival Status')
plt.legend()
plt.show()
```

```python
# Survival rate by gender
gender_survival_rate = titanic_df.groupby('Sex')['Survived'].mean()
print(f'Survival Rate by Gender:\n{gender_survival_rate}\n')
```

```python
# Survival rate by passenger class
class_survival_rate = titanic_df.groupby('Pclass')['Survived'].mean()
print(f'Survival Rate by Passenger Class:\n{class_survival_rate}\n')
```

```python
 # Survival rate by embarked location
embarked_survival_rate = titanic_df.groupby(['Embarked_Q', 'Embarked_S'])['Survived'].mean()
print(f'Survival Rate by Embarked Location:\n{embarked_survival_rate}\n')
```

```python
# Count plot of 'Pclass'
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Pclass', data=titanic_df)
plt.title('Count of Passengers by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
legend_labels = ['Class 1', 'Class 2', 'Class 3']
for i, legend in enumerate(legend_labels):
    ax.bar(i, 0, color=ax.patches[i].get_facecolor(), label=legend)
plt.legend(title='Pclass')
plt.show()
```

```python
#Count plot of 'Sex'
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Sex', data=titanic_df)
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
legend_labels = ['Female', 'Male']
for i, legend in enumerate(legend_labels):
    ax.bar(i, 0, color=ax.patches[i].get_facecolor(), label=legend)
plt.legend(title='Sex')
plt.show()
```

```python
# Heatmap of the correlation matrix
# We need to exclude non-numeric columns for the correlation matrix
numeric_df = titanic_df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
